# Data loading

```python
!pip install pandas
```

```python
import pandas as pd

df = pd.read_csv('/content/Istanbul Weather Data.csv',sep=",")
df.head()
```

```python
df = df[["Rain", "MinTemp", "AvgHumidity", "AvgPressure"]]
df.head()
```

```python
df.shape
```

# Data processing

```python
df = df.dropna()
import numpy as np
labels = np.array(df.pop('Rain'))
features = np.array(df)
```

```python
features
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.15, shuffle=True)
```

# Model

```python
import tensorflow as tf
```

# Building model

```python
NB_classes = 1
NB_neurons = 30
NB_features = 3
activation_func = tf.keras.activations.relu

model = tf.keras.Sequential([
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func, input_shape=(NB_features,)),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dense(NB_neurons,activation=activation_func),
                              tf.keras.layers.Dropout(0.4),
                              tf.keras.layers.Dense(NB_classes)
])

model.compile(optimizer="adam",loss=tf.keras.losses.categorical_crossentropy, metrics=['accuracy'])

model.summary()
```

# Training model

```python
model.fit(x=train_features,
          y=train_labels,
          epochs=30,
          validation_data=(test_features,test_labels),
          verbose=1,
          shuffle=True)
```

```python
performance=model.evaluate(test_features, test_labels, batch_size=32, verbose=1, steps=None, )[1]*100
print('Final accuracy: ', round(performance), '%')
```

# Converting model

# Tensorflow lite

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

open("/content/tflite_model.tflite", "wb").write(tflite_model)
```
